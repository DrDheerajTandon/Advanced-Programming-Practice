# Bar Plot

```python
from matplotlib import pyplot as plt
```

```python
x1 = [2,4,6,8]
y1 = [10,20,30,40]
x2 = [5,10,15,20]
y2 = [100,150,200,250]
plt.bar(x1,y1,label='Line One')
plt.bar(x2,y2,label='Line Tow')
plt.title('Bar Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.show()
```

```python
x1 = [2,4,6,8]
y1 = [10,20,30,40]
x2 = [5,10,15,20]
y2 = [100,150,200,250]
plt.barh(x1,y1,label='Line One')
plt.barh(x2,y2,label='Line Tow')
plt.title('Bar Plot in Horizontal Direction')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.show()
```

# Histogram Plot

```python
l = [2,5,10,15,20,25,28,30,37,40,42,50,55,67,77,80]
plt.hist(l)
plt.title('Histogram Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
l = [2,5,10,15,20,25,28,30,37,40,42,50,55,67,77,80]
plt.hist(l,bins=5,edgecolor='r')
plt.title('Histogram Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

# Scatter Plot

```python
x1 = [10,12,17,21,25,30,34,30,50,60]
y1 = [100,120,130,142,150,160,180,200,220,250]
plt.scatter(x1,y1,color='k',marker='X',s=100)
plt.title('Scatter Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

# Pie Plot

```python
slices = [10,25,45,50,60]
names = ['A','B','C','D','E']
clrs = ['r','c','y','b','m']
plt.pie(slices,
       labels= names,
        colors = clrs,
        shadow=True,
        startangle=120,
        explode=(0.5,0,0,0.6,0),
        autopct='%.2f%%'
       )
plt.title('Pie Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
x1 = [1,5,10,15]
y1 = [5,10,15,20]
x2 = [10,20,30,40]
y2 = [25,50,75,100]
plt.subplot(211)
plt.plot(x1,y1)
plt.subplot(212)
plt.plot(x2,y2)
#plt.title('Multiple Plots using Matplotlib')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

# Functional Programming

```python
import math
def area(r):
    a = math.pi*(r**2)
    return 'The area is {}'.format(a)

area(2)
```

```python
import math
def area(r):
    a = math.pi*(r**2)
    return 'The area is {}'.format(a)

radius = [1,2,3,4,5]
areas = []
for i in radius:
    a1 = area(i)
    areas.append(a1)
    
print(areas)
```
