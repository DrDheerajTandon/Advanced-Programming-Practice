# Style

```python
from matplotlib import style
```

```python
style.use('ggplot')
x = [10,3,6]
y= [4,80,11]
plt.plot(x,y,color='r')
plt.title('Matplotlib Styles')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
style.use('ggplot')
x = [10,3,6]
y= [4,80,11]
plt.plot(x,y,color='g',linewidth=8)
plt.title('Matplotlib Styles')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
style.use('ggplot')
x = [10,3,6]
y= [4,80,11]
plt.plot(x,y,color='g',linewidth=8,label='Line 1')
plt.title('Matplotlib Styles')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
style.use('ggplot')
x1 = [10,3,6]
y1= [4,80,11]
x2 = [10,20,30]
y2 = [50,100,150]
plt.plot(x1,y1,color='g',linewidth=8,label='Line 1')
plt.plot(x2,y2,color='b',linewidth=5,label='Line 2')
plt.title('Matplotlib Styles')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

# Bar Plot

```python
x1 = [1,3,5,7]
y1 = [2,4,6,8]
x2 = [2,6,9,12]
y2 = [4,8,12,16]
plt.bar(x1,y1,label='Label 1',color='g')
plt.bar(x2,y2,label='Label 2',color='r')
plt.title('Bar Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.show()
```
